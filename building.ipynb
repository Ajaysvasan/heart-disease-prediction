```python
import pandas as pd
import numpy as np
import pickle
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
```

```python
data = pd.read_csv('reprocessed.hungarian.data',sep=" ",header=None)
data
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data
```

```python
x = np.array(data.drop(13,axis=1))
y = np.array(data[13])
```

```python
xTrain,xTest,yTrain,yTest = train_test_split(x,y,test_size=0.1)
```

```python
modelOne = LogisticRegression()
modelOne.fit(xTrain,yTrain)
prediction = modelOne.predict(xTest)
acc = accuracy_score(prediction,yTest)
```

```python
print(acc)
```

```python
modelTwo = RandomForestClassifier()
best = 0
for i in range(50):
    xTrain,xTest,yTrain,yTest = train_test_split(x,y,test_size=0.1)
    modelTwo.fit(xTrain,yTrain)
    prediction = modelTwo.predict(xTest)
    acc = accuracy_score(prediction,yTest)
    if acc>best:
        best=acc
        with open("model.pickle",'wb') as file:
            pickle.dump(modelTwo,file)
```

```python
print(best)
```
